```python

import pandas as pd
import matplotlib.pyplot as plt


def explore_link_data(file_path):
    # Load the DataFrame from the CSV file
    df = pd.read_csv(file_path)
    

    character_counts = df['link_title'].value_counts()
    

    top_characters = character_counts.head(10)
    top_characters.plot(kind='bar', color='skyblue')
    plt.title('Top 10 links by Frequency of Appearance')
    plt.xlabel('Character')
    plt.ylabel('Frequency')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

# Define the file path
file_path = '/Users/amimai875/Documents/ds105/DS105W/ds105w-2024-w08-summative-amimai875/task2.csv'

# Call the function to explore link data and create a plot
explore_link_data(file_path)

def explore_link_data(file_path):

    df = pd.read_csv(file_path)
    
    links_per_paragraph = df.groupby('paragraph_id').size()
    

    plt.figure(figsize=(10, 6))
    plt.hist(links_per_paragraph, bins=range(1, links_per_paragraph.max() + 2), color='pink', edgecolor='black')
    plt.title('Distribution of Number of Links per Paragraph')
    plt.xlabel('Number of Links')
    plt.ylabel('Frequency')
    plt.xticks(range(1, links_per_paragraph.max() + 1))
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()

file_path = '/Users/amimai875/Documents/ds105/DS105W/ds105w-2024-w08-summative-amimai875/task2.csv'

explore_link_data(file_path)
```
